```python
from nbformat import read

# Load the Jupyter Notebook
with open('../MachineLearning_Predict_House_Prices.ipynb', 'r', encoding='utf-8') as nb_file:
    nb = read(nb_file, as_version=4)

# Function to create markdown list item with correct indentation
def create_markdown_list_item(level, title):
    indent = '  ' * (level - 1)  # Two spaces per indentation level
    return f"{indent}- {title}"

# Extract the titles from the notebook's cells and format them for Markdown
markdown_list = []
for cell in nb.cells:
    if cell.cell_type == 'markdown':
        lines = cell.source.split('\n')
        for line in lines:
            if line.startswith('#'):
                level = line.count('#')  # Determine the level of the header
                title = line[level:].strip()  # Extract the title text
                markdown_list.append(create_markdown_list_item(level, title))

# Join the list items into a single string
markdown_content = '\n'.join(markdown_list)
print(markdown_content)

```
