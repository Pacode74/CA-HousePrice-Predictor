```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("../housing.csv")
```

## **pandas's Approach**

```python
# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Plotting distributions of key numerical features
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(16, 15))
fig.suptitle('Distributions of Key Numerical Features', fontsize=20)

# Longtude 
sns.histplot(df['longitude'], bins=50, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Longitude')

# Latitude 
sns.histplot(df['latitude'], bins=50, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Latitude')

# Housing Median Age
sns.histplot(df['housing_median_age'], bins=50, kde=True, ax=axes[0, 2])
axes[0, 2].set_title('Housing Median Age')

# Total Rooms
sns.histplot(df['total_rooms'], bins=50, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Total Rooms')

# Total Bedrooms
sns.histplot(df['total_bedrooms'].dropna(), bins=50, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Total Bedrooms')

# Population
sns.histplot(df['population'], bins=50, kde=True, ax=axes[1, 2])
axes[1, 2].set_title('Population')

# Household
sns.histplot(df['households'], bins=50, kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Households')

# Median Income
sns.histplot(df['median_income'], bins=50, kde=True, ax=axes[2, 1])
axes[2, 1].set_title('Median Income')

# Median House Value
sns.histplot(df['median_house_value'], bins=50, kde=True, ax=axes[2, 2])
axes[2, 2].set_title('Median House Value')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust subplots to fit into the figure area.
plt.show()
```

## **Pythonic Approach**

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Define the figure size and subplots
plt.figure(figsize=(20, 15))

# List of columns to plot
columns_to_plot = ['latitude', 'longitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 
                   'population', 'households', 'median_income', 'median_house_value']

# Loop through the list and create a subplot for each column
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[column].dropna(), kde=True, bins=50)  # Drop NA values for 'total_bedrooms'
    plt.title(column.replace('_', ' ').title())

plt.tight_layout()
plt.show()
```
